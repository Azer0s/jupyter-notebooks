### Heads or Tails

$p(Heads) = 0.5$

$p(Tails) = 0.5$

```python
import random
import matplotlib.pyplot as plt

heads_amount = 0
tails_amount = 0
for i in range(1000):
    if random.choice(["heads", "tails"]) == "heads":
        heads_amount += 1
    else:
        tails_amount += 1

plt.bar(["heads", "tails"], [heads_amount, tails_amount])
plt.show()
```

### Test score

1 - 6

```python
result = [0] * 6
for i in range(1000):
    result[random.randrange(0, 6)] += 1

plt.bar([1, 2, 3, 4, 5, 6], result)
plt.show()
```

Probability is NOT EQUAL for each grade

```python
import numpy as np
from scipy.stats import norm 
import statistics 
```

```python
x_axis = np.arange(1, 7, 1.0) 
  
mean = statistics.mean(x_axis) 
sd = statistics.stdev(x_axis)
  
plt.plot(x_axis, norm.pdf(x_axis, mean, sd)) 
plt.show()
```

```python
student_results = [x * 1000 for x in norm.pdf(x_axis, mean, sd)]
plt.bar(x_axis, student_results)
plt.show()
```
