```python
import matplotlib.pyplot as plt

def mean(v):
    return sum(v) / len(v)

def plotXY(vals):
    vx = [x for (x, y) in vals]
    vy = [y for (x, y) in vals]
    plt.plot(vx, vy)
```

$f(x)=kx+d$

$k = \frac{\overline{x}*\overline{y}-\overline{xy}}{(\overline{x})²-\overline{x²}}$

```python
def inferK(xs, ys):
    xsAvg = mean(xs)
    ysAvg = mean(ys)
    xysAvg = mean([x * y for (x, y) in zip(xs, ys)])
    xs2Avg = mean([x * x for x in xs])

    return (xsAvg * ysAvg - xysAvg) / (xsAvg ** 2 - xs2Avg)
```

$d = \overline{y}-k*\overline{x}$

```python
def inferD(k, xs, ys):
    xsAvg = mean(xs)
    ysAvg = mean(ys)

    return ysAvg - k * xsAvg
```

```python
# House price in 1k$ (x)
# Square feet (y)
vals = [(245, 1400), (312, 1600), (279, 1700), (308, 1875), (199, 1100), (219, 1550), (405, 2350), (324, 2450), (319, 1425), (255, 1700)]
vals = [(float(x), float(y)) for (x, y) in vals]
vx = [x for (x, y) in vals]
vy = [y for (x, y) in vals]
```

```python
k = inferK(vx, vy)
d = inferD(k, vx, vy)

minX = min(vx)
maxX = max(vx)

line = [(x, k * x + d) for x in range(int(minX), int(maxX))]

plt.plot(vx, vy, 'o')
plotXY(line)
plt.show()
```
